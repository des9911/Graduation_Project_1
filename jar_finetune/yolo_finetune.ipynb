```python
# train_yolo_pottery_goryeo_joseon.py
from ultralytics import YOLO
from google.colab import drive

# Google Drive 마운트
drive.mount('/content/drive')

# 데이터 경로
data_path = '/content/drive/MyDrive/project/dataset.yaml'

# 1. 사전학습된 YOLOv8 모델 로드
model = YOLO('yolov8s.pt')

print("=== [1단계] YOLOv8s 모델 로드 완료 ===")

# 2. 고려청자 / 조선분청사기 데이터 학습
results = model.train(
    data=data_path,                     # 데이터셋 설정 파일
    epochs=200,                              # 학습 반복 횟수
    imgsz=640,                               # 입력 이미지 크기
    batch=2,                                 # 배치 크기 (VRAM 상황에 맞게 조정)
    lr0=0.001,                               # 초기 학습률
    lrf=0.0001,                              # 최종 학습률
    optimizer='Adam',                        # 옵티마이저
    pretrained=True,                         # 전이학습
    name='goryeo_joseon_pottery_v1',         # 결과 저장 폴더 이름
    project='/content/drive/MyDrive/project/runs/train',                    # 기본 저장 경로
    device=0,                                # GPU 사용 (없으면 'cpu')
    patience=20,                             # early stopping
    augment=True,                            # 데이터 증강
)

print("=== [2단계] 학습 완료 ===")

# 3. 검증 단계
metrics = model.val()
print("=== [3단계] 검증 완료 ===")
print(metrics)

# 4. 예시 이미지 예측
test_result = model.predict(
    source='/content/drive/MyDrive/project/test_images/sample2.jpg',
    save=True,
    project='/content/drive/MyDrive/project/predict_results',  # 저장할 폴더
    name='exp1'  # 하위 폴더 이름
)
print("=== [4단계] 예측 완료 ===")
print("결과는 runs/predict 폴더에 저장됩니다.")
```

```python
# 2. Í≤ΩÎ°ú / ÏÉÅÏàò ÏÑ§Ï†ï

# Python: ÌîÑÎ°úÏ†ùÌä∏ Í≤ΩÎ°úÏôÄ Ï£ºÏöî ÏÉÅÏàò ÏÑ§Ï†ï (ÌïÑÏöî Ïãú Í≤ΩÎ°ú ÏàòÏ†ï)
import os, random, numpy as np, torch
PROJECT_ROOT = '/content/drive/MyDrive/yolo&clip'  # ÌïÑÏöîÏãú Î≥ÄÍ≤Ω
YOLO_DATA_DIR = os.path.join(PROJECT_ROOT, 'yolo_dataset')   # Label Studio Í≤∞Í≥º ÏúÑÏπò (images + labels)
CROPPED_DIR = os.path.join(PROJECT_ROOT, 'cropped')         # YOLOÎ°ú cropÎêú Ïù¥ÎØ∏ÏßÄ Ï†ÄÏû• ÏúÑÏπò
CLIP_DATA_DIR = os.path.join(PROJECT_ROOT, 'clip_dataset')  # CLIP ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ Ìè¥Îçî
MODELS_DIR = os.path.join(PROJECT_ROOT, 'models')           # Î™®Îç∏ Ï†ÄÏû• Ìè¥Îçî
YOLO_RUNS_DIR = os.path.join(PROJECT_ROOT, 'runs')     # ultralytics runs Ï†ÄÏû• ÏúÑÏπò
os.makedirs(CROPPED_DIR, exist_ok=True)
os.makedirs(MODELS_DIR, exist_ok=True)
os.makedirs(YOLO_RUNS_DIR, exist_ok=True)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", DEVICE)
# ÌÅ¥ÎûòÏä§ ÏÑ§Ï†ï (ÏöîÍµ¨ Î∞òÏòÅ)
PARENT_CLASSES = ['celadon', 'white_porcelain', 'other_pottery']
SUB_PROMPTS = {
    'celadon': [["Goryeo celadon", "Buncheong ware"],
                ["Goryeo dynasty inlaid celadon pottery with jade green glaze, Korean celadon bowl with sanggam inlay technique",
                 "Joseon dynasty buncheong ware with stamped decoration, traditional Korean buncheong bowl with grey-green glaze"]],
    'white_porcelain': [["Joseon white porcelain", "blue-painted porcelain", "iron-painted porcelain", "bronze painted porcelain"],
                        ["Joseon dynasty white porcelain jar, unpainted, smooth white glaze, traditional Korean baekja",
                         "blue and white porcelain from the Joseon dynasty",
                         "Joseon dynasty iron-painted white porcelain with black painted decoration, traditional Korean jar or bowl",
                         "Joseon dynasty polychrome painted (red color painted) porcelain called dongchae baekja"]],
    'other_pottery': [["other pottery"], ["Korean earthenware pottery that is not celadon or white porcelain"]]
}

# ÌïôÏäµ ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ (ÌïÑÏöîÏãú Ï°∞Ï†ï)
YOLO_EPOCHS = 50
YOLO_BATCH = 16
YOLO_IMGSZ = 640

CLIP_1ST_EPOCHS = 20
CLIP_BATCH = 8
LR = 1e-4
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
```

```python
# 3. YOLO dataset yaml ÏÉùÏÑ± Î∞è ÎùºÎ≤® Ìè¨Îß∑ ÏÑ§Î™Ö

# YOLO Îç∞Ïù¥ÌÑ∞ Íµ¨ÏÑ± ÌôïÏù∏ Î∞è yaml ÏûëÏÑ± (ultralyticsÏóêÏÑú ÏÇ¨Ïö©)
yolo_yaml = os.path.join(PROJECT_ROOT, 'yolo_dataset.yaml')
content = f"""
train: {YOLO_DATA_DIR}/images/train
val:   {YOLO_DATA_DIR}/images/val

nc: 1
names: ['pottery']
"""
os.makedirs(os.path.join(YOLO_DATA_DIR,'images','train'), exist_ok=True)
os.makedirs(os.path.join(YOLO_DATA_DIR,'images','val'), exist_ok=True)
os.makedirs(os.path.join(YOLO_DATA_DIR,'labels','train'), exist_ok=True)
os.makedirs(os.path.join(YOLO_DATA_DIR,'labels','val'), exist_ok=True)
with open(yolo_yaml, 'w') as f:
    f.write(content.strip())
print("YOLO yaml saved:", yolo_yaml)

print("\nÎùºÎ≤® Ìè¨Îß∑(Í∞Å Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÏùëÌïòÎäî .txt):")
print("ÌÅ¥ÎûòÏä§ x_center y_center width height  (Î™®Îëê Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞Î°ú Ï†ïÍ∑úÌôî, ÌÅ¥ÎûòÏä§ id = 0 -> pottery)")
```

```python
# 4. YOLO ÌïôÏäµ (runs ÎîîÎ†âÌÑ∞Î¶¨ Î≥ÄÍ≤Ω)

# YOLO ÌïôÏäµ (ultralytics)
from ultralytics import YOLO
import os
#os.environ['YOLO_RUNS_DIR'] = YOLO_RUNS_DIR  # ultralyticsÏóêÏÑú runs Ï†ÄÏû• ÏúÑÏπòÎ°ú ÏÇ¨Ïö©

# ÏÇ¨Ï†ÑÌïôÏäµÎêú weight Î°úÎìú (yolov8n Í∞ÄÎ≤ºÏõÄ)
yolo_model = YOLO('yolov8n.pt')

# ÌïôÏäµ ÏàòÌñâ (Ï£ºÏùò: Îç∞Ïù¥ÌÑ∞Í∞Ä Ï§ÄÎπÑÎêòÏñ¥ ÏûàÏñ¥Ïïº Ìï®)
print("Starting YOLO training... (this may take time)")
yolo_model.train(data=yolo_yaml, epochs=YOLO_EPOCHS, imgsz=YOLO_IMGSZ, batch=YOLO_BATCH, project=YOLO_RUNS_DIR, name='pottery_detection')
print("YOLO training finished. Check runs directory:", YOLO_RUNS_DIR)
```

```python
# 5. YOLO Ï∂îÎ°† -> ÏûêÎèô Crop Ï†ÄÏû• Ìï®Ïàò

# YOLO Ï∂îÎ°† ÌõÑ crop ÏûêÎèô Ï†ÄÏû• (ÌïôÏäµÎêú best.pt Í≤ΩÎ°úÎ•º ÏûêÎèô ÌÉêÏÉâ)
import glob, cv2, os
from ultralytics import YOLO

# ÌïôÏäµ Í≤∞Í≥º Í≤ΩÎ°úÏóêÏÑú best.pt Ï∞æÍ∏∞
candidates = glob.glob(os.path.join(YOLO_RUNS_DIR, 'pottery_detection', 'weights', 'best.pt'))
if len(candidates)>0:
    yolo_weights = candidates[-1]
    print("Found YOLO trained weights:", yolo_weights)
    yolo_inf = YOLO(yolo_weights)
else:
    print("No trained YOLO weights found in runs. Using pretrained yolov8n for inference.")
    yolo_inf = YOLO('yolov8n.pt')

def yolo_crop_and_save(image_path, save_dir=CROPPED_DIR, prefix='crop'):
    os.makedirs(save_dir, exist_ok=True)
    results = yolo_inf.predict(source=image_path, imgsz=YOLO_IMGSZ, conf=0.25)  # conf threshold Ï°∞Ï†ï Í∞ÄÎä•
    crops=[]
    img = cv2.imread(image_path)
    if img is None:
        return crops
    res = results[0]
    if not hasattr(res, 'boxes') or len(res.boxes)==0:
        return crops
    for i, box in enumerate(res.boxes.xyxy):
        x1,y1,x2,y2 = map(int, box.cpu().numpy())
        x1,y1 = max(0,x1), max(0,y1)
        x2,y2 = min(img.shape[1]-1,x2), min(img.shape[0]-1,y2)
        crop = img[y1:y2, x1:x2]
        if crop.size==0: continue
        fname = f"{prefix}_{os.path.basename(image_path).split('.')[0]}_{i}.jpg"
        outp = os.path.join(save_dir, fname)
        cv2.imwrite(outp, crop)
        crops.append(outp)
    return crops

# YOLO Îç∞Ïù¥ÌÑ∞Ïùò val Ìè¥Îçî ÎÇ¥ Ïù¥ÎØ∏ÏßÄ crop
sample_images = glob.glob(os.path.join(YOLO_DATA_DIR, 'images', 'val', '*'))
if len(sample_images)>0:
    for i in range(len(sample_images)):
      s = sample_images[i]
      print("Sample image crop:", s)
      print(yolo_crop_and_save(s))
else:
    print("No sample images in YOLO val. You can run yolo_crop_and_save on any image path.")

# YOLO Îç∞Ïù¥ÌÑ∞Ïùò train Ìè¥Îçî ÎÇ¥ Ïù¥ÎØ∏ÏßÄ crop
sample_images = glob.glob(os.path.join(YOLO_DATA_DIR, 'images', 'train', '*'))
if len(sample_images)>0:
    for i in range(len(sample_images)):
      s = sample_images[i]
      print("Sample image crop:", s)
      print(yolo_crop_and_save(s))
else:
    print("No sample images in YOLO val. You can run yolo_crop_and_save on any image path.")
```

```python
# 6. CROPPED -> CLIP ÌïôÏäµÏö© ÏûêÎèô Ï†ïÎ¶¨ (ÌååÏùºÎ™Ö ÌÇ§ÏõåÎìú Í∏∞Î∞ò)

# cropped Ìè¥ÎçîÏùò ÌååÏùºÎ™ÖÏóê ÌÇ§ÏõåÎìú(Ïòà: goryeo, buncheong, joseon, blue, iron, bronze, other Îì±)Í∞Ä ÏûàÏúºÎ©¥ ÏûêÎèôÏúºÎ°ú CLIP train/val ÎîîÎ†âÌÑ∞Î¶¨Ïóê Î∂ÑÎ•ò
import shutil, random, glob

keyword_to_parent = {
    'buncheong':'celadon',
    'goryeo':'celadon',
    'joseon':'white_porcelain',
    'blue':'white_porcelain',
    'iron':'white_porcelain',
    'bronze':'white_porcelain',
    'other':'other_pottery'
}

# CLIP Îç∞Ïù¥ÌÑ∞ Ìè¥Îçî Íµ¨Ï°∞ ÏÉùÏÑ±
for split in ['train','val']:
    for c in PARENT_CLASSES:
        os.makedirs(os.path.join(CLIP_DATA_DIR, split, c), exist_ok=True)

def auto_organize_crops(source_dir=CROPPED_DIR, split_ratio=0.85):
    files = glob.glob(os.path.join(source_dir, '*.jpg')) + glob.glob(os.path.join(source_dir, '*.png'))
    random.shuffle(files)
    for f in files:
        fname = os.path.basename(f).lower()
        assigned=None
        for kw, parent in keyword_to_parent.items():
            if kw in fname:
                assigned = parent
                break
        if assigned is None:
            # default: other_pottery (ÎòêÎäî ÏàòÎèôÍ≤ÄÌÜ† Í∂åÏû•)
            assigned = 'other_pottery'
        # split
        dest_split = 'train' if random.random() < split_ratio else 'val'
        dest_dir = os.path.join(CLIP_DATA_DIR, dest_split, assigned)
        shutil.copy(f, os.path.join(dest_dir, fname))
    print("Auto organizing done. Check:", CLIP_DATA_DIR)

# Ïã§Ìñâ (Ï£ºÏùò: ÌååÏùºÎ™ÖÏóê ÌÇ§ÏõåÎìúÍ∞Ä ÏûàÏñ¥Ïïº ÏûêÎèô Î∂ÑÎ•òÍ∞Ä Ïûò Îê®)
auto_organize_crops()
```

```python
# 7. CLIP 1Ï∞® ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞Î°úÎçî (Ï¶ùÍ∞ï Ìè¨Ìï®)

# CLIP Î™®Îç∏ Î°úÎìúÏôÄ DataLoader Íµ¨ÏÑ± (albumentations ÏÇ¨Ïö©)
from transformers import CLIPProcessor, CLIPModel
from torch.utils.data import Dataset, DataLoader, ConcatDataset
from PIL import Image
import albumentations as A
from albumentations.pytorch import ToTensorV2

processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch16")
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch16").to(DEVICE)

# augmentation
train_tf = A.Compose([
    A.RandomResizedCrop((224,224), scale=(0.8,1.0)),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.4),
    A.GaussNoise(p=0.2),
    A.ShiftScaleRotate(p=0.4),
    A.ColorJitter(p=0.4),
    A.CLAHE(p=0.3),
    A.MotionBlur(p=0.2),
    A.Normalize(mean=(0.48145466,0.4578275,0.40821073), std=(0.26862954,0.26130258,0.27577711)),
    ToTensorV2()
])
val_tf = A.Compose([
    A.Resize(224,224),
    A.Normalize(mean=(0.48145466,0.4578275,0.40821073), std=(0.26862954,0.26130258,0.27577711)),
    ToTensorV2()
])

class ParentDataset(Dataset):
    def __init__(self, root, split, classes, transform):
        self.samples=[]; self.labels=[]
        self.transform = transform
        for i, c in enumerate(classes):
            folder = os.path.join(root, split, c)
            if not os.path.isdir(folder): continue
            for f in os.listdir(folder):
                if f.lower().endswith(('.jpg','.png','.jpeg')):
                    self.samples.append(os.path.join(folder,f))
                    self.labels.append(i)
    def __len__(self): return len(self.samples)
    def __getitem__(self, idx):
        p = self.samples[idx]
        img = Image.open(p).convert('RGB')
        img = np.array(img)
        img = self.transform(image=img)['image']
        label = torch.tensor(self.labels[idx], dtype=torch.long)
        return img, label

# build loaders
train_ds = ParentDataset(CLIP_DATA_DIR, 'train', PARENT_CLASSES, train_tf)
val_ds = ParentDataset(CLIP_DATA_DIR, 'val', PARENT_CLASSES, val_tf)
print("Parent train samples:", len(train_ds), "val samples:", len(val_ds))
train_loader = DataLoader(train_ds, batch_size=CLIP_BATCH, shuffle=True, num_workers=2)
val_loader = DataLoader(val_ds, batch_size=CLIP_BATCH, shuffle=False, num_workers=2)

test_ds = ConcatDataset([train_ds, val_ds])
test_loader = DataLoader(test_ds, batch_size=CLIP_BATCH, shuffle=False, num_workers=2)
```

```python
# 8. CLIP 1Ï∞® Î∂ÑÎ•òÍ∏∞ Ï†ïÏùò (vision frozen) Î∞è ÌïôÏäµ

# parent classifier: CLIP vision feature -> FC head
import torch.nn as nn, torch.optim as optim
vision_dim = clip_model.vision_model.config.hidden_size
class ParentClassifier(nn.Module):
    def __init__(self, vision_dim, n_classes):
        super().__init__()
        self.head = nn.Sequential(
            nn.Linear(vision_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, n_classes)
        )
    def forward(self, pixel_values):
        # pixel_values: [B, C, H, W]
        vision_out = clip_model.vision_model(pixel_values=pixel_values).pooler_output  # [B, vision_dim]
        return self.head(vision_out)

parent_classifier = ParentClassifier(vision_dim, len(PARENT_CLASSES)).to(DEVICE)

# freeze CLIP vision encoder (Í∂åÏû•: Îç∞Ïù¥ÌÑ∞ Ï†ÅÏùÑ Îïå)
for p in clip_model.parameters():
    p.requires_grad = False

optimizer = optim.AdamW(parent_classifier.parameters(), lr=LR)
criterion = nn.CrossEntropyLoss()

# ÌïôÏäµ Î£®ÌîÑ
from tqdm import tqdm
best_val = 0.0
for epoch in range(CLIP_1ST_EPOCHS):
    parent_classifier.train()
    running_loss=0.0; all_preds=[]; all_labels=[]
    for imgs, labels in tqdm(train_loader, desc=f"Train ep{epoch+1}"):
        imgs = imgs.to(DEVICE).float()
        labels = labels.to(DEVICE)
        logits = parent_classifier(imgs)
        loss = criterion(logits, labels)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        running_loss += loss.item()*imgs.size(0)
        all_preds.extend(logits.argmax(dim=1).cpu().numpy().tolist())
        all_labels.extend(labels.cpu().numpy().tolist())
    train_acc = (np.array(all_preds)==np.array(all_labels)).mean() if len(all_labels)>0 else 0.0
    print(f"Epoch {epoch+1} Train Loss:{running_loss/len(train_ds):.4f} Acc:{train_acc:.4f}")

    # val
    parent_classifier.eval()
    vpred=[]; vtrue=[]
    with torch.no_grad():
        for imgs, labels in val_loader:
            imgs = imgs.to(DEVICE).float(); labels = labels.to(DEVICE)
            logits = parent_classifier(imgs)
            vpred.extend(logits.argmax(dim=1).cpu().numpy().tolist())
            vtrue.extend(labels.cpu().numpy().tolist())
    if len(vtrue)>0:
        val_acc = (np.array(vpred)==np.array(vtrue)).mean()
        print(f"Val Acc: {val_acc:.4f}")
        if val_acc>best_val:
            best_val=val_acc
            torch.save(parent_classifier.state_dict(), os.path.join(MODELS_DIR, 'clip_parent_best.pth'))
            print("Saved best parent classifier.")
    else:
        print("No validation data to evaluate.")
```

```python
# 9. Evaluation (confusion matrix, classification report)

# ÌèâÍ∞Ä Î∞è ÏãúÍ∞ÅÌôî
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# load best
parent_classifier.load_state_dict(torch.load(os.path.join(MODELS_DIR, 'clip_parent_best.pth'), map_location=DEVICE))
parent_classifier.eval()

# evaluate on val or train if val empty
eval_loader = val_loader if len(val_ds)>0 else train_loader
y_true=[]; y_pred=[]
with torch.no_grad():
    for imgs, labels in eval_loader:
        imgs = imgs.to(DEVICE).float()
        logits = parent_classifier(imgs)
        y_pred.extend(logits.argmax(dim=1).cpu().numpy().tolist())
        y_true.extend(labels.numpy().tolist())

print(classification_report(y_true, y_pred, target_names=PARENT_CLASSES))
cm = confusion_matrix(y_true, y_pred, labels=list(range(len(PARENT_CLASSES))))
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=PARENT_CLASSES, yticklabels=PARENT_CLASSES, cmap='Blues')
plt.xlabel('Predicted'); plt.ylabel('True'); plt.title('Parent-level Confusion Matrix'); plt.show()

# evaluate on total
eval_loader = test_loader
y_true=[]; y_pred=[]
with torch.no_grad():
    for imgs, labels in eval_loader:
        imgs = imgs.to(DEVICE).float()
        logits = parent_classifier(imgs)
        y_pred.extend(logits.argmax(dim=1).cpu().numpy().tolist())
        y_true.extend(labels.numpy().tolist())

print(classification_report(y_true, y_pred, target_names=PARENT_CLASSES))
cm = confusion_matrix(y_true, y_pred, labels=list(range(len(PARENT_CLASSES))))
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=PARENT_CLASSES, yticklabels=PARENT_CLASSES, cmap='Blues')
plt.xlabel('Predicted'); plt.ylabel('True'); plt.title('Parent-level Confusion Matrix'); plt.show()
```

```python
# 10. CLIP 2Ï∞® Zero-shot Ìï®Ïàò Ï†ïÏùò

# Zero-shot Î∂ÑÎ•ò: Ï£ºÏñ¥ÏßÑ crop(PIL image)Í≥º parentÏóê ÎåÄÌï¥ ÌõÑÎ≥¥ ÌÖçÏä§Ìä∏ÏôÄ cosine similarity Í≥ÑÏÇ∞
from PIL import Image

def clip_zero_shot_subclass(pil_image, parent_label):
    texts = SUB_PROMPTS[parent_label][1]
    # processor handles tokenization and image preprocessing
    img_inputs = processor(images=pil_image, return_tensors="pt").pixel_values.to(DEVICE)
    txt_inputs = processor(text=texts, return_tensors="pt", padding=True)
    # ÌÖçÏä§Ìä∏ Ïù∏ÌíãÏùÄ tokenizer->input_ids; transformer's CLIPModel has get_text_features
    input_ids = txt_inputs['input_ids'].to(DEVICE)
    attention_mask = txt_inputs['attention_mask'].to(DEVICE)
    with torch.no_grad():
        img_feat = clip_model.get_image_features(img_inputs)   # [1, dim]
        txt_feat = clip_model.get_text_features(input_ids=input_ids, attention_mask=attention_mask)  # [n, dim]
        img_feat = img_feat / img_feat.norm(dim=-1, keepdim=True)
        txt_feat = txt_feat / txt_feat.norm(dim=-1, keepdim=True)
        sims = (img_feat @ txt_feat.T).squeeze(0).cpu().numpy()  # [n]
    idx = int(np.argmax(sims))
    return SUB_PROMPTS[parent_label][0][idx], float(sims[idx]), sims.tolist()
```

```python
# 11. ÌÜµÌï© Ï∂îÎ°† Ìï®Ïàò (YOLO -> CLIP parent -> CLIP zero-shot sub)

# ÌÜµÌï© ÌååÏù¥ÌîÑÎùºÏù∏: Ïù¥ÎØ∏ÏßÄ ÌïòÎÇò ÎÑ£ÏúºÎ©¥ YOLO detect -> crops -> parent predict -> sub zero-shot -> annotate Ï†ÄÏû•
import cv2
from PIL import Image
def infer_single_image(image_path, out_path=None, conf=0.25):
    results = yolo_inf.predict(source=image_path, imgsz=YOLO_IMGSZ, conf=conf)
    img = cv2.imread(image_path)
    annotated = img.copy()
    outputs=[]
    if len(results)==0: return outputs, annotated
    for i, box in enumerate(results[0].boxes.xyxy):
        x1,y1,x2,y2 = map(int, box.cpu().numpy())
        crop = img[y1:y2, x1:x2]
        if crop.size==0: continue
        pil = Image.fromarray(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)).convert('RGB')
        # parent predict
        # use val_transform to normalize, then forward
        import numpy as np
        temp = val_tf(image=np.array(pil))['image'].unsqueeze(0).to(DEVICE).float()
        with torch.no_grad():
            plog = parent_classifier(temp)
            pidx = int(plog.argmax(dim=1).item())
            pscore = float(torch.softmax(plog, dim=1)[0,pidx].cpu().item())
            parent_label = PARENT_CLASSES[pidx]
        # sub zero-shot
        sub_text, sub_score, all_scores = clip_zero_shot_subclass(pil, parent_label)
        outputs.append({'bbox':(x1,y1,x2,y2), 'parent':parent_label, 'parent_score':pscore, 'sub_text':sub_text, 'sub_score':sub_score})
        label = f"{parent_label}({pscore:.2f})"
        second_label = f"{sub_text}({sub_score:.2f})"
        cv2.rectangle(annotated, (x1,y1),(x2,y2), (0,255,0), 2)
        cv2.putText(annotated, label, (x1, max(15,y1-7)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1)
        (text_w, text_h), baseline = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)
        cv2.putText(annotated, second_label, (x1, y1 - 7 - text_h - 2), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1)
    if out_path:
        cv2.imwrite(out_path, annotated)
    return outputs, annotated

# ÏÇ¨Ïö© Ïòà:
out, ann = infer_single_image(os.path.join(PROJECT_ROOT,'test','final_test.png'), out_path=os.path.join(PROJECT_ROOT, 'test', 'annotated_final_test.png'))
from matplotlib import pyplot as plt
plt.imshow(cv2.cvtColor(ann, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

```python
# 13. prompt Î≥ÄÍ≤ΩÏóê Îî∞Î•∏ Í≤∞Í≥º Ï∞®Ïù¥ ÌôïÏù∏

# -------------------------------------------------------------------
# infer_compare_prompts_once : Îã®Ïùº Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÌï¥ prompt ÏÑ∏Ìä∏ ÎπÑÍµê Ïã§Ìñâ
# Ï∂úÎ†•: ÏΩòÏÜî(ÌîÑÎ°¨ÌîÑÌä∏Î≥Ñ score Ìëú) + matplotlibÏúºÎ°ú annotated Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
# -------------------------------------------------------------------
import os, cv2, torch, numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# ÏïàÏ†ÑÏû•Ïπò: ÌïÑÏöîÌïú Î≥ÄÏàòÎì§Ïù¥ ÏûàÎÇò Í∞ÑÎã® Ï≤¥ÌÅ¨ (ÏóÜÏúºÎ©¥ ÏóêÎü¨Î©îÏãúÏßÄ)
required = ['yolo_inf','parent_classifier','val_tf','clip_model','processor','PARENT_CLASSES','SUB_PROMPTS','DEVICE','YOLO_IMGSZ']
missing = [v for v in required if v not in globals()]
if missing:
    print("‚ö†Ô∏è ÏïÑÎûò Ï†ÑÏó≠ Î≥ÄÏàòÍ∞Ä Ï†ïÏùòÎêòÏñ¥ ÏûàÏñ¥Ïïº Ìï©ÎãàÎã§:", missing)
    # Ìï®Ïàò Ï†ïÏùòÎäî Í≥ÑÏÜçÌïòÏßÄÎßå Ïã§Ìñâ Ï†ÑÏóê Î∞òÎìúÏãú Ï†ïÏùòÌï¥Ïïº Ìï®

# helper: Ï£ºÏñ¥ÏßÑ prompt Î¶¨Ïä§Ìä∏Ïóê ÎåÄÌï¥ CLIP zero-shot ÏàòÌñâ (Ïù¥ÎØ∏ÏßÄ: PIL.Image)
def zero_shot_with_prompts(pil_image, prompt_list):
    """
    pil_image : PIL.Image (RGB)
    prompt_list: [str, str, ...]  -> Í∞Å ÌîÑÎ°¨ÌîÑÌä∏Ïóê ÎåÄÌïú ÌÖçÏä§Ìä∏(ÏôÑÏ†Ñ Î¨∏Ïû• Í∞ÄÎä•)
    returns: (best_text, best_score, scores_list (in same order as prompt_list))
    """
    # prepare inputs
    img_inputs = processor(images=pil_image, return_tensors="pt").pixel_values.to(DEVICE)  # [1, C, H, W]
    # process texts using processor -> get tokenized input_ids/attention_mask
    txt_inputs = processor(text=prompt_list, return_tensors="pt", padding=True)
    input_ids = txt_inputs['input_ids'].to(DEVICE)
    attention_mask = txt_inputs['attention_mask'].to(DEVICE)

    with torch.no_grad():
        img_feat = clip_model.get_image_features(img_inputs)            # [1, D]
        txt_feat = clip_model.get_text_features(input_ids=input_ids, attention_mask=attention_mask)  # [N, D]
        # normalize
        img_feat = img_feat / img_feat.norm(dim=-1, keepdim=True)
        txt_feat = txt_feat / txt_feat.norm(dim=-1, keepdim=True)
        sims = (img_feat @ txt_feat.T).squeeze(0).cpu().numpy()  # [N]
    best_idx = int(np.argmax(sims))
    return prompt_list[best_idx], float(sims[best_idx]), sims.tolist()

# Î©îÏù∏ Ìï®Ïàò: Îã®Ïùº Ïù¥ÎØ∏ÏßÄ Ïã§ÌñâÏö©
def infer_compare_prompts_once(image_path, conf=0.25, show_image=True):
    """
    1) YOLO detect -> crops
    2) Í∞Å bboxÏóê ÎåÄÌï¥ parent predict
    3) parentÏóê ÎßûÎäî SUB_PROMPTS[parent] ÎÇ¥ Í∞Å prompt-set(Î¶¨Ïä§Ìä∏)ÏùÑ ÏàúÌöåÌïòÎ©∞ zero-shot (Í∞Å prompt-set ÎÇ¥ Ïó¨Îü¨ ÌÖçÏä§Ìä∏ Ï§ë best ÏÑ†ÌÉù)
    4) ÏΩòÏÜîÏóê Ìëú Ïä§ÌÉÄÏùºÎ°ú Ï∂úÎ†•. Ïù¥ÎØ∏ÏßÄÏóêÎäî parent + best_overall_sub (ÏµúÍ≥† Ï†êÏàò)Îßå ÌëúÏãú
    """
    if not os.path.isfile(image_path):
        raise FileNotFoundError(f"Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§: {image_path}")

    # 1) YOLO detect
    results = yolo_inf.predict(source=image_path, imgsz=YOLO_IMGSZ, conf=conf)
    img_bgr = cv2.imread(image_path)
    annotated = img_bgr.copy()
    outputs = []

    if len(results) == 0 or not hasattr(results[0], 'boxes') or len(results[0].boxes) == 0:
        print("‚ö†Ô∏è ÌÉêÏßÄÎêú Í∞ùÏ≤¥Í∞Ä ÏóÜÏäµÎãàÎã§.")
        if show_image:
            plt.figure(figsize=(8,8)); plt.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)); plt.axis('off'); plt.show()
        return outputs, annotated

    # print header
    num_boxes = len(results[0].boxes)
    print(f"[ DETECTED {num_boxes} OBJECT(S) ]\n")

    for idx, box in enumerate(results[0].boxes.xyxy.cpu().numpy()):
        x1, y1, x2, y2 = map(int, box)
        crop = img_bgr[y1:y2, x1:x2]
        if crop.size == 0:
            continue
        pil = Image.fromarray(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)).convert('RGB')

        # 2) parent predict (val_tf -> tensor -> classifier)
        try:
            temp = val_tf(image=np.array(pil))['image'].unsqueeze(0).to(DEVICE).float()
        except Exception as e:
            # fallback: if val_tf is a torchvision transform (Tensor output)
            try:
                temp = val_tf(pil).unsqueeze(0).to(DEVICE).float()
            except Exception as e2:
                raise RuntimeError("val_tf transform Ï≤òÎ¶¨ Ïò§Î•ò. val_tfÍ∞Ä PIL->tensor Î≥ÄÌôòÏùÑ ÏàòÌñâÌï¥Ïïº Ìï©ÎãàÎã§.") from e2

        with torch.no_grad():
            plog = parent_classifier(temp)
            pidx = int(plog.argmax(dim=1).item())
            pscore = float(torch.softmax(plog, dim=1)[0, pidx].cpu().item())
            parent_label = PARENT_CLASSES[pidx]

        # 3) SUB_PROMPTS[parent_label]ÏóêÎäî prompt-set Î¶¨Ïä§Ìä∏Í∞Ä Îì§Ïñ¥ÏûàÏùå
        #    Ïòà: SUB_PROMPTS['celadon'] == [ [promptA1, promptA2], [promptB1, promptB2] ]
        prompt_sets = PROMPTS_SET.get(parent_label, [])
        prompt_results = []  # list of dicts per prompt set

        for ps_idx, prompt_list in enumerate(prompt_sets):
            # prompt_list is a list of strings (candidate texts)
            best_text, best_score, all_scores = zero_shot_with_prompts(pil, prompt_list)
            prompt_results.append({
                'prompt_set_idx': ps_idx,
                'prompt_list': prompt_list,
                'best_text': best_text,
                'best_score': best_score,
                'all_scores': all_scores
            })

        # determine overall best among prompt_sets (compare best_score)
        if len(prompt_results) > 0:
            best_overall = max(prompt_results, key=lambda x: x['best_score'])
            best_text_overall = best_overall['best_text']
            best_score_overall = best_overall['best_score']
        else:
            best_text_overall, best_score_overall = "N/A", 0.0

        # store outputs
        outputs.append({
            'bbox': (x1, y1, x2, y2),
            'parent': parent_label,
            'parent_score': pscore,
            'prompt_results': prompt_results,
            'best_sub_text': best_text_overall,
            'best_sub_score': best_score_overall
        })

        # -------------------
        # ÏΩòÏÜî Ï∂úÎ†•: Î≥¥Í∏∞Ï¢ãÍ≤å Ï†ïÎ†¨
        # -------------------
        print(f"#{idx+1} bbox={x1,y1,x2,y2} | parent={parent_label} ({pscore:.3f})")
        print("-" * 60)
        for pr in prompt_results:
            pset_no = pr['prompt_set_idx'] + 1
            # print prompt-set ÎåÄÌëú(Í∞ÑÎã®Ìôî)
            print(f"[Prompt set {pset_no}] candidates:")
            for cand, s in zip(pr['prompt_list'], pr['all_scores']):
                print(f"  - {cand}  ‚Üí  score: {s:.4f}")
            print(f"  => best: {pr['best_text']}  (score: {pr['best_score']:.4f})\n")
        print(">> Best overall zero-shot:", best_text_overall, f"({best_score_overall:.4f})")
        print("="*60 + "\n")

        # Ïù¥ÎØ∏ÏßÄ annotate: parent + best overall sub (ÏòµÏÖò B)
        label_main = f"{parent_label} ({pscore:.2f})"
        label_sub = f"{best_text_overall} ({best_score_overall:.2f})"
        # draw bbox
        cv2.rectangle(annotated, (x1, y1), (x2, y2), (0, 200, 0), 2)
        # put parent label (below top-left)
        cv2.putText(annotated, label_main, (x1, y2 + 18), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,200,0), 1, cv2.LINE_AA)
        # put best sub label (below parent)
        cv2.putText(annotated, label_sub, (x1, y2 + 36), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (0,200,0), 1, cv2.LINE_AA)

    # show annotated image
    if show_image:
        plt.figure(figsize=(8,8))
        plt.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()

    return outputs, annotated

# -------------------------------
# ÏÇ¨Ïö© ÏòàÏãú (Îã®Ïùº Ïù¥ÎØ∏ÏßÄ):
# -------------------------------
img_path = os.path.join(PROJECT_ROOT, 'test', 'final_test.png')
outputs, annotated_img = infer_compare_prompts_once(img_path, conf=0.25, show_image=True)
print("Done. outputs:", outputs)
```

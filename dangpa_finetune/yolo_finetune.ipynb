```python
from ultralytics import YOLO

# 1. 모델 로드
model = YOLO('yolov8s.pt') 

print("학습 시작")

# 2. 파인튜닝 실행 
results = model.train(
    data='dataset.yaml',         
    epochs=200,                  
    imgsz=640,                   # 이미지 크기
    batch=4,                     # 배치 크기 (로컬 VRAM에 맞춰 
    lr0=0.001,                   # 초기 학습률 (낮게 설정하여 전이 학습
    lrf=0.0001,                  # 최종 학습률
    name='local_dangpa_verifier_model_notebook'  # 결과 출력 폴더
)

print("--- 학습 완료 ---")
```
